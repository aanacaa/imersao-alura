## Script para gerar a amostra
Disponibilizado no treinamento da Alura (adaptado)

```
import pandas as pd
import numpy as np
import random


%precision %.2f
pd.options.display.float_format = '{:,.2f}'.format
```

```
from google.colab import drive
drive.mount('/content/drive/')
```

```
ano = "2019"
fonte = f"/content/drive/My Drive/Colab Notebooks/Imersao-Alura/MICRODADOS_ENEM_{ano}.csv"

seed = 43278
np.random.seed(seed)

num_linhas = sum(1 for l in open(fonte,encoding="ISO-8859-1"))

p = 0.025
tamanho = int(num_linhas * p)

index = random.sample(range(1, num_linhas), num_linhas - tamanho)

dados = pd.read_csv(fonte, 
                    encoding="ISO-8859-1",
                    delimiter=";",
                    skiprows=index)
dados.shape
```
